```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score
from decision_tree import DecisionTree
```

```python
df_train = pd.read_csv("../data/churn_data.csv", index_col=[0])
X = df_train.drop(["Exited"], axis=1)
y = df_train[["Exited"]]
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)
```

```python
tree = DecisionTree()
tree.fit(X, y)
y_pred = tree.predict(X)
```

```python
print(f"Accuracy: {round(accuracy_score(y, y_pred), 3)}")
print(f"Precision: {round(precision_score(y, y_pred), 3)}")
print(f"Recall: {round(recall_score(y, y_pred), 3)}")
```
