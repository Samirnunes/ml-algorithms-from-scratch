```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from som import SOM
```

```python
df = pd.read_csv("../data/adm_data.csv", index_col=[0])
X = df.drop(["Chance of Admit "], axis=1)
y = df[["Chance of Admit "]]
X_orig = X.copy()
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)
```

```python
nx=2
ny=2
learning_rate = 0.1
neighborhood_radius = 1
decay_ration = 1.001
tol = 0.0001
max_iter = 1500
random_state = 0
som = SOM(nx=nx, ny=ny, learning_rate=learning_rate, neighborhood_radius=neighborhood_radius, 
          decay_ratio = decay_ration, tol=tol, max_iter=max_iter, random_state=random_state)
som.fit(X)
```
